# B. What is the average price of an item sold at each branch of the city

```python
np.mean(df['Unit price'])
```

```python
average_price = df.groupby('Product line')['Unit price'].mean()
average_price
```

```python
df['average_price'] = df.groupby('Product line')['Unit price'].transform('mean')
df.head()
```

```python
sns.barplot(df['Branch'],df['average_price'],hue=df['City'])
```

```python
sns.barplot(df['Branch'],df['average_price'],hue=df['Product line'])
```

```python
sns.boxplot(df['City'],df['average_price'],hue=df['Branch'])
```

```python
#here i will do feature construction 
# Create a new column combining city and branch information
df['City_Branch'] = df['City'] + '_' + df['Branch']

# Calculate average price of an item sold at each city branch
average_pricesbranchwise = df.groupby('City_Branch')['Unit price'].mean()

average_pricesbranchwise
```

```python
df['average_pricesbranchwise'] = df.groupby('City_Branch')['Unit price'].transform('mean')
df.head()
```

```python
sns.barplot(x=df['City_Branch'], y=df['average_pricesbranchwise'], data=df)
plt.title('Distribution of Average Prices Across City Branches')
plt.xlabel('City Branch')
plt.ylabel('Average Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid for better visualization
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

# C. Analyze the performance of sales and revenue, Month over Month across theProduct line, Gender, and Payment Method, and identify the focus areas toget better sales for April 2019.

```python
df['Sales']=df['Product line'].map(Sales)
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
#months
df['month']=df['Date'].dt.month_name()
df.head(10)
```

sales declined in the month of march

```python
df['month'].unique()
```

```python
# Create a list of months in the desired order
desired_order = ['January', 'February', 'March']

# Convert 'month' column to categorical with desired order
df['month'] = pd.Categorical(df['month'], categories=desired_order, ordered=True)

# Plot the line plot with the specified order
sns.lineplot(data=df, x='month', y='Sales')

# Show the plot
plt.show()
```

sales increased in january and peaked in Feburary but after that it gradually declined to march

```python

sns.lineplot(data=df, x='month', y='Revenue')
```

```python
sns.lineplot(data=df, x='month', y='Revenue', markers=True,dashes=False,ci=None,label='Revenue')
sns.lineplot(data=df, x='month', y='Sales', markers=True,dashes=False,ci=None,label='Sales')
plt.figure(figsize=(10, 6))
```
